1. ** What was the maximum total revenue among all the shops in September, 2014?** 


* Hereinafter *revenue* refers to total sales minus value of goods returned.

```python
transactions.head()

total_sale = transactions['item_price']*transactions['item_cnt_day']

transactions['total_sale'] = total_sale

transactions.head()
```

```python
# september is 9th month
transactions['date'] = pd.to_datetime(transactions['date'])
```

```python
transactions.set_index(['date'], inplace= True)
maskSep = transactions.loc['2014-09-01':'2014-09-30']
maskSep
```

```python
maskSep.groupby(['shop_id']).sum()['total_sale']
```

* Note, that for an object `x` of type `pd.Series`: `x.argmax()` returns **index** of the maximum element. `pd.Series` can have non-trivial index (not `[1, 2, 3, ... ]`).

```python
# jone is 6th and Aug is 8th month
maskSummer = transactions.loc['2014-06-01':'2014-09-30']
maskSummer
```

```python
maxRevID = maskSummer.groupby(['item_id']).sum()['total_sale'].argmax()
maxRevID
```
